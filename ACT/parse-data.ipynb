```python
image_dst = './dataset/source_image/'
mask_dst = './dataset/segmentation_image/'

for base_folder in glob.glob('./others/demo/demo*'):
    images = glob.glob(base_folder + '/img/*')
    for image_path in images:
        image_name = image_path.split("/")[-1].split(".")[0] # no extension.
        annotation_path = base_folder + '/ann/' + image_name + '.json'
        image_dst_name = image_dst + image_path.split("/")[-1]
        mask_path = mask_dst + image_name + '.png'
        
        with open(annotation_path) as f:
            anno = json.load(f)

        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        """
        construct mask from json.
        """
        sanity_check = [1 for obj in anno['objects'] if obj['points']['exterior']]
        if len(anno['objects']) == len(sanity_check): # all object with exterior shape.
            res = np.zeros_like(img)
            for obj in anno['objects']:
                exterior = obj['points']['exterior']
                poly = np.array(exterior, dtype=np.int32)
                mask = np.zeros((img.shape[0], img.shape[1]))
                print("current image: ", image_name)
                cv2.fillPoly(mask, [poly], 255)
                person = mask
                for interior in obj['points']['interior']:
                    poly = np.array(interior, dtype=np.int32)
                    mask = np.zeros((img.shape[0], img.shape[1]))
                    cv2.fillConvexPoly(mask, poly, 255)
                    person = person - mask
                res = res + person
            cv2.imwrite(mask_path, res) # write mask file to dst.
            copyfile(image_path, image_dst_name) # copy source image to dst.
        else:
            print("this img doesn't contain experior coordinates points.")
```

```python
"""
Convert mask into color-indexed images. 
"""
import tensorflow as tf
from PIL import Image
from tqdm import tqdm
import numpy as np

import os, shutil

# palette (color map) describes the (R, G, B): Label pair
palette = {0 : 0 ,
           255 : 1
         }

def convert_from_color_segmentation(arr):
    arr_2d = np.zeros((arr.shape[0], arr.shape[1]), dtype=np.uint8)

    for c, i in palette.items():
        m = np.all(arr == c)
        arr_2d[m] = i
    return arr_2d


label_dir = './dataset/segmentation_image/'
new_label_dir = './dataset/new_segmentation_image/'

if not os.path.isdir(new_label_dir):
    print("creating folder: ",new_label_dir)
    os.mkdir(new_label_dir)
else:
    print("Folder alread exists. Delete the folder and re-run the code!!!")


label_files = os.listdir(label_dir)

for l_f in tqdm(label_files):
    arr = np.array(Image.open(label_dir + l_f))
    arr_2d = convert_from_color_segmentation(arr)
    Image.fromarray(arr_2d).save(new_label_dir + l_f)
```
